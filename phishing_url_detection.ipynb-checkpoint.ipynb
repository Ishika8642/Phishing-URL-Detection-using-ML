## 3. Preprocessing and Feature Engineering

We select useful features, convert categorical data to numerical if needed, and prepare our data for training.


```python
import re

def extract_url_features(df):
    df['url_length'] = df['url'].apply(lambda x: len(str(x)))
    df['num_dots'] = df['url'].apply(lambda x: str(x).count('.'))
    df['has_https'] = df['url'].apply(lambda x: 1 if 'https' in str(x) else 0)
    df['num_hyphens'] = df['url'].apply(lambda x: str(x).count('-'))
    df['num_slashes'] = df['url'].apply(lambda x: str(x).count('/'))
    df['has_at_symbol'] = df['url'].apply(lambda x: 1 if '@' in str(x) else 0)

    # Check if URL uses IP address instead of domain
    def has_ip(url):
        try:
            return 1 if re.match(r"https?://(\d{1,3}\.){3}\d{1,3}", str(url)) else 0
        except:
            return 0
    df['has_ip'] = df['url'].apply(has_ip)

    # Check for common shortener services
    def is_shortened(url):
        shorteners = ['bit.ly', 'tinyurl', 'goo.gl', 'ow.ly', 't.co']
        return 1 if any(s in str(url) for s in shorteners) else 0
    df['is_shortened'] = df['url'].apply(is_shortened)

    return df
```

```python
# 2. Extract features from URL
train_df = extract_url_features(train_df)
test_df = extract_url_features(test_df)

# 3. Prepare training data
X_train = train_df.drop(['status', 'url'], axis=1)
y_train = train_df['status']

X_test = test_df.drop(['status', 'url'], axis=1)
y_test = test_df['status']
```

## 4. Model Training

We use a Random Forest Classifier â€” a powerful and interpretable ensemble method â€” to train on the data.


```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

## 5. Model Evaluation

We evaluate our model using accuracy, precision, recall, F1 score, and visualize performance using a confusion matrix and ROC curve.


## 6. Visualizing Results

We plot feature importance to understand what features the model relies on most.


```python
from sklearn.metrics import roc_curve, auc

# Predict probabilities for the 'phishing' class
y_prob = model.predict_proba(X_test)[:, list(model.classes_).index('phishing')]

# Compute ROC curve and AUC with correct positive label
fpr, tpr, thresholds = roc_curve(y_test, y_prob, pos_label='phishing')
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ðŸ§ª ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
